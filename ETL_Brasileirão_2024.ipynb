### **Criação da tabela do campeonato**

```python
rodada = 0
campeonato = []
for i in jogos:
  rodada += 1
  for j in i:
    jogo = j.split(" x ")
    linha = [rodada, jogo[0][:-1].strip(), int(jogo[0][-1]), jogo[1][1:].strip(), int(jogo[1][0])]
    campeonato.append(linha)
campeonato
```
